```python
import pandas as pd
```

```python
zoo = pd.read_csv('zoo.csv', index_col = False, names = ['animal_name', 'hair', 'feathers','eggs','milk','airborne','aquatic','predator','toothed','backbone','breathes','venomous','fins','legs','tail','domestic', 'catsize', 'type'])
```

```python
type_name = {1:'mammal', 2:'bird', 3:'reptile', 4:'fish', 5:'amphibian', 6:'bug', 7:'invertebrae'}
```

```python
zoo['type_name'] = zoo.type.map(type_name)
```

```python
zoo
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test = tts(zoo[['hair', 'feathers','eggs','milk','airborne','aquatic','predator','toothed','backbone','breathes','venomous','fins','legs','tail','domestic', 'catsize']], zoo['type'], test_size = 0.05, shuffle = False)
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier as dtc
```

```python
classifier = dtc()
```

```python
classifier.fit(X_train, y_train)
```

```python
classifier.predict(X_test)
```
